### **UNIVARIATE ANALYSIS**

```python
sns.histplot(x='price', data=df);
```

```python
# univariate distribution of the price feature
sns.histplot(df['price']);
```

- I observed that the distribution of the car price feature is positively skewed.
- Greater number of cars from our data are between 5000 - 20,000

```python
# Univariate distribution of the car fuel types
sns.countplot(x='fueltype', data=df);
```

- Greater number of Cars from our data set uses gas other than diesel
- Diesel car types are relatively low

```python
# Univariate distribution of engine type
sns.countplot(x='enginetype', data=df)
```

- Ohc engine is the highiest used engine type in cars from our dataset

```python
# Univariate distribution of car body
sns.countplot(x='carbody', data=df)
```

- Sedan type is the highest carbody from our data set followed by hatchback
- Convertible is the lowest carbody from our data set followed by hardtop

```python
# Univariate distribution for stroke
sns.histplot(x='stroke', data=df);
```

- The stroke distribution is negatively skewed

### **BIVARIATE ANALYSIS**


```python
df.info()
```

```python
# Relationship of cars Horsepower and price
sns.jointplot(x='horsepower', y='price', data=df)
```

- I observed a positive correlation between the horsepower feature and the dependent variable
- horsepower feature is fairly symetrical, though I can detect outliers

```python
sns.jointplot(x='horsepower', y='price', data=df, kind='scatter')
```

```python
# Relationship between cars engine location and price
sns.boxplot(x='enginelocation', y='price', data=df);
```

- I can observe that on average less expensive cars have their engines in the front of the car, though I detect the presence of outliers
- I can observe that all cars that have engines located at the rear axis are on average more expensive than those cars that have engine in the front axis

```python
# Relationship between engine location and engine size
sns.barplot(x='enginelocation', y='enginesize', data=df);
```

- i observe from the data set that Cars with big engine size have their engine located at the rear compared to the front engine location

```python
# Relationship between engine type and highwaympg
sns.stripplot(x='enginetype', y='highwaympg', data=df)
```

- I can observe that the ohc engine on average has a higher mpg

```python
# Relationship between the Car body and Price
sns.boxplot(x='carbody', y= 'price', data=df, palette='magma');
```

- I can observe that on average hardtop and convertible cars tends to be more expensive.
